# Модуль 12: Паттерны проектирования

## 1. Что такое паттерны проектирования

Паттерны проектирования — это проверенные подходы к решению типовых задач, которые регулярно возникают при разработке программных систем.

Они описывают не готовую реализацию, а общую идею организации кода:
- как распределить ответственность между классами и объектами
- как сделать решение расширяемым
- как уменьшить связанность и упростить сопровождение

## 2. Причины возникновения паттернов проектирования

Паттерны появились потому, что в разных проектах снова и снова встречаются одни и те же архитектурные трудности.

Основные причины их возникновения:
- повторяемость типовых проблем при разработке
- необходимость делать крупные системы понятными и поддерживаемыми
- потребность в общем профессиональном языке для обсуждения архитектуры
- накопление успешных практик, которые можно переиспользовать

## 3. Понятие паттерна проектирования

Паттерн проектирования — это формализованное описание типовой архитектурной ситуации, проблемы и способа её решения.

Обычно паттерн включает:
- контекст применения
- проблему, которую он решает
- состав участников и их связи
- последствия использования

То есть паттерн отвечает не только на вопрос "что сделать", но и "когда это уместно".

## 4. Принципы применения паттернов проектирования

Паттерны нужно применять осмысленно, а не автоматически.

Базовые принципы такие:
- использовать паттерн только при наличии реальной архитектурной проблемы
- не превращать паттерн в самоцель или украшение кода
- учитывать цену усложнения архитектуры
- проверять, делает ли решение код гибче, понятнее и удобнее для сопровождения
- адаптировать идею паттерна под задачу, а не копировать её механически

## 5. Принципы выбора паттернов проектирования

Выбор паттерна начинается не с названия, а с анализа проблемы.

Ориентироваться стоит на такие принципы:
- сначала определить, где именно возникает сложность: в создании объектов, в структуре или в поведении
- выбирать самый простой паттерн, который действительно закрывает задачу
- учитывать масштаб проекта и вероятность будущих изменений
- смотреть, насколько решение будет понятно команде
- сравнивать выгоду от гибкости с дополнительной сложностью

Практическое правило:
- сначала делай простой код
- добавляй паттерн только тогда, когда он убирает реальную сложность

## 6. Принципы разделения паттернов на категории

Классическое разделение паттернов строится по характеру решаемой задачи:
- порождающие паттерны управляют созданием объектов
- структурные паттерны описывают способы композиции классов и объектов
- поведенческие паттерны организуют взаимодействие, алгоритмы и распределение ответственности

Такое деление помогает быстрее находить подходящее решение и сравнивать паттерны внутри одной группы.

## 7. Введение в UML

UML (Unified Modeling Language) — это язык визуального описания программных систем.

При изучении паттернов особенно важны три вида диаграмм:
- диаграмма классов показывает классы, интерфейсы, связи и роли участников паттерна
- диаграмма объектов показывает конкретные экземпляры объектов и их состояние в определённый момент времени
- диаграмма взаимодействия показывает, как объекты обмениваются сообщениями и в каком порядке выполняются действия

## 8. Использование UML при анализе паттернов проектирования

UML помогает разбирать паттерн не только на уровне идеи, но и на уровне конкретной структуры и взаимодействий.

На практике это выглядит так:
- диаграмма классов помогает увидеть состав участников и связи между ними
- диаграмма объектов позволяет понять, какие экземпляры реально существуют в системе
- диаграмма взаимодействия показывает последовательность вызовов и распределение ответственности

Поэтому UML удобно использовать для объяснения паттернов, сравнения похожих решений и проектирования собственной архитектуры.

## 9. Порождающие паттерны

Порождающие паттерны нужны, когда создание объекта — это уже отдельная задача.

Они полезны, если:
- объект может создаваться по разным правилам
- клиент не должен знать конкретный класс
- объект сложный и собирается по шагам
- выгоднее клонировать существующий экземпляр, чем создавать новый с нуля

### Singleton

**Что это:** паттерн, который гарантирует, что у класса будет только один экземпляр.

**Как работает:** класс хранит ссылку на уже созданный объект и при следующем создании возвращает его же.

**Когда применять:** для общего конфига, логгера, реестра. Использовать осторожно, потому что это форма глобального состояния.

```python
class AppConfig:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
            cls._instance.mode = "production"
        return cls._instance


config1 = AppConfig()
config2 = AppConfig()
print(config1 is config2)
print(config2.mode)
```

### Factory Method

**Что это:** паттерн, который выносит создание объекта в отдельный метод или фабрику.

**Как работает:** клиент просит создать объект по общему интерфейсу, а фабрика решает, какой класс вернуть.

**Когда применять:** когда есть несколько вариантов объектов с одинаковым поведением, например разные каналы уведомлений.

```python
class EmailSender:
    def send(self, message):
        return f"Email: {message}"


class SmsSender:
    def send(self, message):
        return f"SMS: {message}"


class NotificationFactory:
    @staticmethod
    def create(channel):
        if channel == "email":
            return EmailSender()
        if channel == "sms":
            return SmsSender()
        raise ValueError("Unknown channel")


sender = NotificationFactory.create("sms")
print(sender.send("Ваш код подтверждения: 1234"))
```

### Abstract Factory

**Что это:** паттерн для создания семейства связанных объектов.

**Как работает:** одна фабрика создаёт набор совместимых объектов. Например, светлая тема создаёт светлую кнопку и светлое поле ввода, а тёмная — тёмные версии тех же элементов.

**Когда применять:** когда нужно создавать наборы объектов, которые должны использоваться вместе.

```python
class LightButton:
    def render(self):
        return "LightButton"


class LightInput:
    def render(self):
        return "LightInput"


class DarkButton:
    def render(self):
        return "DarkButton"


class DarkInput:
    def render(self):
        return "DarkInput"


class LightThemeFactory:
    def create_button(self):
        return LightButton()

    def create_input(self):
        return LightInput()


class DarkThemeFactory:
    def create_button(self):
        return DarkButton()

    def create_input(self):
        return DarkInput()


def build_form(factory):
    button = factory.create_button()
    field = factory.create_input()
    print(button.render(), field.render())


build_form(LightThemeFactory())
build_form(DarkThemeFactory())
```

### Builder

**Что это:** паттерн для пошаговой сборки сложного объекта.

**Как работает:** объект настраивается шаг за шагом, а затем возвращается готовый результат.

**Когда применять:** когда у объекта много опциональных параметров и длинный конструктор становится неудобным.

```python
class LaptopBuilder:
    def __init__(self):
        self.laptop = {
            "cpu": "Intel i5",
            "ram": 8,
            "ssd": 256,
            "gpu": "integrated",
        }

    def for_study(self):
        self.laptop["ram"] = 16
        self.laptop["ssd"] = 512
        return self

    def for_gaming(self):
        self.laptop["ram"] = 32
        self.laptop["ssd"] = 1024
        self.laptop["gpu"] = "RTX 4070"
        return self

    def with_cpu(self, cpu):
        self.laptop["cpu"] = cpu
        return self

    def build(self):
        return self.laptop.copy()


print(LaptopBuilder().for_study().with_cpu("Apple M3").build())
```

### Prototype

**Что это:** паттерн, который создаёт новый объект копированием существующего.

**Как работает:** берётся готовый шаблон и клонируется, после чего при необходимости меняются отдельные поля.

**Когда применять:** когда объект дорого или неудобно собирать с нуля.

```python
import copy

template_order = {
    "delivery": "standard",
    "gift_wrap": False,
    "items": ["book"],
}

fast_order = copy.deepcopy(template_order)
fast_order["delivery"] = "express"

print(template_order)
print(fast_order)
```

## 10. Структурные паттерны

Структурные паттерны нужны, когда объекты уже существуют, но их нужно грамотно соединить.

Они полезны, если:
- интерфейсы несовместимы
- систему нужно скрыть за более простым API
- поведение хочется добавлять композиционно
- объектную структуру нужно собирать как дерево
- доступ к объекту надо контролировать

### Adapter

**Что это:** паттерн, который преобразует один интерфейс в другой.

**Как работает:** адаптер принимает старый объект и показывает наружу новый ожидаемый интерфейс.

**Когда применять:** для legacy-кода или сторонней библиотеки с неудобным API.

```python
class OldSmsService:
    def send_sms(self, phone, text):
        print(f"old service -> {phone}: {text}")


class SmsAdapter:
    def __init__(self, service, phone):
        self.service = service
        self.phone = phone

    def send(self, message):
        self.service.send_sms(self.phone, message)


SmsAdapter(OldSmsService(), "+79990001122").send("Курьер уже в пути")
```

### Bridge

**Что это:** паттерн, который разделяет абстракцию и реализацию, чтобы их можно было менять независимо.

**Как работает:** один объект хранит ссылку на реализацию и делегирует ей работу.

**Когда применять:** когда есть две независимые оси изменений. Например, разные типы пультов и разные устройства.

```python
class TV:
    def turn_on(self):
        return "TV is on"


class Radio:
    def turn_on(self):
        return "Radio is on"


class RemoteControl:
    def __init__(self, device):
        self.device = device

    def power(self):
        return self.device.turn_on()


print(RemoteControl(TV()).power())
print(RemoteControl(Radio()).power())
```

### Composite

**Что это:** паттерн, который позволяет работать с группой объектов так же, как с одним объектом.

**Как работает:** и отдельный элемент, и контейнер имеют общий интерфейс. Контейнер хранит вложенные элементы и делегирует вызовы им.

**Когда применять:** для древовидных структур, например папки и файлы, меню и подпункты.

```python
class File:
    def __init__(self, name, size):
        self.name = name
        self.size = size

    def get_size(self):
        return self.size


class Folder:
    def __init__(self, name):
        self.name = name
        self.children = []

    def add(self, child):
        self.children.append(child)

    def get_size(self):
        return sum(child.get_size() for child in self.children)


docs = Folder("docs")
docs.add(File("readme.txt", 10))
docs.add(File("report.pdf", 90))
print(docs.get_size())
```

### Decorator

**Что это:** паттерн, который добавляет объекту новое поведение без изменения его класса.

**Как работает:** декоратор оборачивает объект и повторяет тот же интерфейс, добавляя свою логику до или после вызова.

**Когда применять:** когда нужно гибко комбинировать дополнительные обязанности.

```python
class Coffee:
    def price(self):
        return 120

    def description(self):
        return "Кофе"


class MilkDecorator:
    def __init__(self, drink):
        self.drink = drink

    def price(self):
        return self.drink.price() + 30

    def description(self):
        return self.drink.description() + ", молоко"


class SyrupDecorator:
    def __init__(self, drink):
        self.drink = drink

    def price(self):
        return self.drink.price() + 25

    def description(self):
        return self.drink.description() + ", сироп"


drink = SyrupDecorator(MilkDecorator(Coffee()))
print(drink.description())
print(drink.price())
```

### Facade

**Что это:** паттерн, который даёт простой интерфейс к сложной подсистеме.

**Как работает:** фасад знает, какие внутренние сервисы и в каком порядке нужно вызвать.

**Когда применять:** когда нужно скрыть сложность внутренней системы за одним понятным методом.

```python
class PaymentService:
    def pay(self, amount):
        print(f"Оплата {amount} подтверждена")


class WarehouseService:
    def reserve(self, item):
        print(f"Товар '{item}' зарезервирован")


class DeliveryService:
    def create(self, item):
        print(f"Доставка для '{item}' создана")


class OrderFacade:
    def __init__(self):
        self.payment = PaymentService()
        self.warehouse = WarehouseService()
        self.delivery = DeliveryService()

    def place_order(self, item, amount):
        self.payment.pay(amount)
        self.warehouse.reserve(item)
        self.delivery.create(item)
        print("Заказ оформлен")


OrderFacade().place_order("Наушники", 7990)
```

### Flyweight

**Что это:** паттерн, который выносит общую повторяющуюся часть состояния в разделяемые объекты.

**Как работает:** тяжёлая общая информация создаётся один раз и переиспользуется многими объектами.

**Когда применять:** когда в системе много похожих мелких объектов и важно экономить память.

```python
class TreeType:
    def __init__(self, name, color):
        self.name = name
        self.color = color


class TreeFactory:
    _types = {}

    @classmethod
    def get_type(cls, name, color):
        key = (name, color)
        if key not in cls._types:
            cls._types[key] = TreeType(name, color)
        return cls._types[key]


class Tree:
    def __init__(self, x, y, tree_type):
        self.x = x
        self.y = y
        self.tree_type = tree_type


pine = TreeFactory.get_type("pine", "green")
forest = [Tree(10, 20, pine), Tree(30, 40, pine)]
print(forest[0].tree_type is forest[1].tree_type)
```

### Proxy

**Что это:** паттерн, в котором объект-заместитель контролирует доступ к реальному объекту.

**Как работает:** proxy имеет тот же интерфейс, что и реальный объект, но может добавлять ленивую загрузку, проверку прав или кэширование.

**Когда применять:** когда реальный объект дорогой, удалённый или требует контроля доступа.

```python
class Image:
    def __init__(self, path):
        print("Heavy image loading")
        self.path = path

    def show(self):
        print(f"Show {self.path}")


class ImageProxy:
    def __init__(self, path):
        self.path = path
        self._real = None

    def show(self):
        if self._real is None:
            self._real = Image(self.path)
        self._real.show()


img = ImageProxy("photo.png")
img.show()
img.show()
```

## 11. Поведенческие паттерны

Поведенческие паттерны нужны, когда основная сложность связана с логикой, алгоритмами и взаимодействием объектов.

Они полезны, если:
- нужно переключать алгоритмы
- один объект должен уведомлять многих
- поведение зависит от состояния
- запрос должен проходить через цепочку обработчиков
- хочется вынести действие в отдельный объект

### Chain of Responsibility

**Что это:** паттерн, в котором запрос проходит по цепочке обработчиков.

**Как работает:** каждый обработчик либо решает задачу, либо передаёт её дальше.

**Когда применять:** когда заранее неизвестно, кто именно должен обработать запрос.

```python
class Handler:
    def __init__(self, next_handler=None):
        self.next_handler = next_handler

    def handle(self, request):
        if self.next_handler:
            return self.next_handler.handle(request)
        return "Unhandled"


class AuthHandler(Handler):
    def handle(self, request):
        if not request.get("user"):
            return "401 Unauthorized"
        return super().handle(request)


class RoleHandler(Handler):
    def handle(self, request):
        if request.get("role") != "admin":
            return "403 Forbidden"
        return "200 OK"


chain = AuthHandler(RoleHandler())
print(chain.handle({"user": "alice", "role": "admin"}))
print(chain.handle({"user": "bob", "role": "user"}))
```

### Command

**Что это:** паттерн, который превращает действие в отдельный объект.

**Как работает:** команда знает, что и над каким объектом нужно выполнить. Благодаря этому команды можно складывать в очередь, логировать или отменять.

**Когда применять:** для кнопок интерфейса, очередей задач, undo/redo.

```python
class Light:
    def on(self):
        print("Light is on")


class TurnOnCommand:
    def __init__(self, light):
        self.light = light

    def execute(self):
        self.light.on()


class Button:
    def __init__(self, command):
        self.command = command

    def press(self):
        self.command.execute()


Button(TurnOnCommand(Light())).press()
```

### Iterator

**Что это:** паттерн, который позволяет последовательно обходить коллекцию, не раскрывая её внутреннее устройство.

**Как работает:** коллекция отдаёт объект-итератор, который знает, как переходить к следующему элементу.

**Когда применять:** когда нужно дать единый способ обхода пользовательской структуры данных.

```python
class BookShelf:
    def __init__(self, books):
        self.books = books

    def __iter__(self):
        return BookShelfIterator(self.books)


class BookShelfIterator:
    def __init__(self, books):
        self.books = books
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.books):
            raise StopIteration
        book = self.books[self.index]
        self.index += 1
        return book


for book in BookShelf(["Clean Code", "Fluent Python", "Refactoring"]):
    print(book)
```

### Mediator

**Что это:** паттерн, который централизует взаимодействие между объектами через посредника.

**Как работает:** участники не общаются напрямую друг с другом, а отправляют сообщения медиатору.

**Когда применять:** когда связи между объектами становятся слишком запутанными.

```python
class ChatMediator:
    def send(self, message, user):
        for colleague in user.colleagues:
            if colleague is not user:
                colleague.receive(message)


class User:
    def __init__(self, name, mediator):
        self.name = name
        self.mediator = mediator
        self.colleagues = []

    def send(self, message):
        self.mediator.send(f"{self.name}: {message}", self)

    def receive(self, message):
        print(message)


mediator = ChatMediator()
alice = User("Alice", mediator)
bob = User("Bob", mediator)
alice.colleagues = [alice, bob]
bob.colleagues = [alice, bob]
alice.send("Привет")
```

### Memento

**Что это:** паттерн, который позволяет сохранять и восстанавливать состояние объекта.

**Как работает:** объект создаёт снимок своего состояния, а внешний менеджер хранит эти снимки.

**Когда применять:** для undo/redo, истории изменений, возврата к прошлой версии.

```python
class Editor:
    def __init__(self):
        self.text = ""

    def write(self, text):
        self.text += text

    def save(self):
        return self.text

    def restore(self, snapshot):
        self.text = snapshot


editor = Editor()
editor.write("Hello")
snapshot = editor.save()
editor.write(", world")
print(editor.text)
editor.restore(snapshot)
print(editor.text)
```

### Observer

**Что это:** паттерн, в котором один объект уведомляет подписчиков о своих изменениях.

**Как работает:** подписчики регистрируются у издателя, а при событии он вызывает их всех.

**Когда применять:** для событийной архитектуры, уведомлений, реакций нескольких подсистем на одно событие.

```python
class Order:
    def __init__(self):
        self.subscribers = []

    def subscribe(self, listener):
        self.subscribers.append(listener)

    def set_status(self, status):
        for listener in self.subscribers:
            listener(status)


def email_listener(status):
    print(f"Email: статус заказа -> {status}")


def crm_listener(status):
    print(f"CRM: обновлён статус -> {status}")


order = Order()
order.subscribe(email_listener)
order.subscribe(crm_listener)
order.set_status("Передан в доставку")
```

### State

**Что это:** паттерн, который выносит поведение для разных состояний объекта в отдельные классы.

**Как работает:** основной объект делегирует действие текущему состоянию, а состояние может перевести объект в следующий режим.

**Когда применять:** когда поведение сильно зависит от текущего этапа жизненного цикла объекта.

```python
class DraftState:
    def publish(self, document):
        document.state = ReviewState()
        return "Черновик отправлен на проверку"


class ReviewState:
    def publish(self, document):
        document.state = PublishedState()
        return "Документ опубликован"


class PublishedState:
    def publish(self, document):
        return "Документ уже опубликован"


class Document:
    def __init__(self):
        self.state = DraftState()

    def publish(self):
        return self.state.publish(self)


doc = Document()
print(doc.publish())
print(doc.publish())
print(doc.publish())
```

### Strategy

**Что это:** паттерн, который выносит взаимозаменяемые алгоритмы в отдельные объекты.

**Как работает:** контекст получает стратегию извне и вызывает её через общий интерфейс.

**Когда применять:** когда есть несколько способов решить одну и ту же задачу.

```python
class StandardDelivery:
    def calculate(self, weight):
        return 200 + weight * 10


class ExpressDelivery:
    def calculate(self, weight):
        return 500 + weight * 20


class PickupDelivery:
    def calculate(self, weight):
        return 0


class DeliveryCalculator:
    def __init__(self, strategy):
        self.strategy = strategy

    def get_price(self, weight):
        return self.strategy.calculate(weight)


print(DeliveryCalculator(StandardDelivery()).get_price(3))
print(DeliveryCalculator(ExpressDelivery()).get_price(3))
print(DeliveryCalculator(PickupDelivery()).get_price(3))
```

### Template Method

**Что это:** паттерн, в котором базовый класс задаёт общий каркас алгоритма, а отдельные шаги переопределяются в наследниках.

**Как работает:** порядок шагов фиксирован в базовом классе, но конкретная реализация частей алгоритма меняется в дочерних классах.

**Когда применять:** когда алгоритм в целом одинаковый, а отдельные шаги различаются.

```python
class ReportTemplate:
    def build(self):
        data = self.fetch_data()
        return self.format_data(data)

    def fetch_data(self):
        raise NotImplementedError

    def format_data(self, data):
        raise NotImplementedError


class SalesReport(ReportTemplate):
    def fetch_data(self):
        return [100, 200, 300]

    def format_data(self, data):
        return f"Сумма продаж: {sum(data)}"


print(SalesReport().build())
```

### Visitor

**Что это:** паттерн, который позволяет добавлять новые операции к объектной структуре без изменения самих элементов.

**Как работает:** каждый элемент принимает visitor и передаёт управление в подходящий метод visitor.

**Когда применять:** когда структура объектов стабильна, а операций над ней становится всё больше.

```python
class Book:
    def __init__(self, title, price):
        self.title = title
        self.price = price

    def accept(self, visitor):
        return visitor.visit_book(self)


class Course:
    def __init__(self, title, price):
        self.title = title
        self.price = price

    def accept(self, visitor):
        return visitor.visit_course(self)


class DiscountVisitor:
    def visit_book(self, book):
        return f"{book.title}: {book.price * 0.9}"

    def visit_course(self, course):
        return f"{course.title}: {course.price * 0.8}"


visitor = DiscountVisitor()
items = [Book("Python Book", 1000), Course("Python Course", 5000)]
for item in items:
    print(item.accept(visitor))
```
